Locating a district and making a separate dataset of it

```python
bogra = df1.loc[df1["District"] == "Bogra"]
bogra
```

Replacing the 1234s with month names

```python
replace_values = {1:"January", 2:"February", 3:"March", 4:"April", 5:"May",
                  6:"June", 7:"July", 8:"August", 9:"September", 10:"October",
                  11:"November", 12:"December"}

bogra["Month\Date"] = bogra["Month\Date"].replace(replace_values)
bogra
```

## Dealing With Nul Values

```python
# replacing "****" with NaN
bogra.replace("****", np.nan, inplace = True)
bogra
```

```python
# evaluating missing data (Boolian value)
missing_data = bogra.isnull()
missing_data
```

## Ploting all the months togather

```python
plt.plot(february['Day'], february['Tempareture'], color= 'green', label= 'February')
plt.plot(march['Day'], march['Tempareture'], color= 'blue', label= 'March')
plt.plot(april['Day'], april['Tempareture'], color= 'red', label= 'April')
plt.plot(may['Day'], may['Tempareture'], color= 'yellow', label= 'May')
plt.plot(june['Day'], june['Tempareture'], color= 'orange', label= 'June')
plt.plot(july['Day'], july['Tempareture'], color= 'black', label= 'July')
plt.plot(august['Day'], august['Tempareture'], color= 'purple', label= 'August')
plt.plot(september['Day'], september['Tempareture'], color= 'orchid', label= 'September')
plt.plot(october['Day'], october['Tempareture'], color= 'palevioletred', label= 'October')
plt.plot(november['Day'], november['Tempareture'], color= 'teal', label= 'November')
plt.plot(december['Day'], december['Tempareture'], color= 'violet', label= 'December')
plt.title("Tempareture of Bogra in 1948")
plt.xticks(['5', '10','15', '20','25','30'])
plt.xlabel('Day')
plt.ylabel('Tempareture')
plt.legend()
```
